```python
import pandas as pd
import plotly.express as px
import panel as pn
```

```python
# variables for filter button sizes
filter_controls_width = 200
filter_width = filter_controls_width - 20
```

```python
# have to load df first because the the pn.depends decorator needs to reference a global df.  Should look for a better way to do this
pn.extension('plotly','tabulator')


df = pd.read_csv('data/heart_failure_clinical_records_dataset.csv')

display(df)
```

```python
# variables are age, anaemia, creatinine_phosphokinase, diabetes, ejection_fraction, high_blood_pressure, platelets, serum_creatinine, serum_sodium, sex, smoking, time, and DEATH_EVENT
# target variable is DEATH_EVENT

numeric_cols = [
    'age',
    'creatinine_phosphokinase',
    'ejection_fraction',
    'platelets',
    'serum_creatinine',
    'serum_sodium',
    'time'
]
```

```python
# update categorical columns to be strings instead of integers
df['DEATH_EVENT_LABEL'] = df['DEATH_EVENT'].map({0: 'Survived', 1: 'Died'})
df['sex_label'] = df['sex'].map({0: 'Female', 1: 'Male'})
df['smoking_label'] = df['smoking'].map({0: 'Non-Smoker', 1: 'Smoker'})
df['diabetes_label'] = df['diabetes'].map({0: 'Non-Diabetic', 1: 'Diabetic'})
df['high_blood_pressure_label'] = df['high_blood_pressure'].map({0: 'No High Blood Pressure', 1: 'High Blood Pressure'})

```

```python
display(df)
```

# Create Widgets

```python
sex_filter = pn.widgets.Select(
    name='Sex Filter',
    options=['All', 'Male', 'Female'],
    value='All',
    width=filter_width
)

outcome_filter = pn.widgets.Select(
    name='Death Outcome Filter',
    options=['All', 'Survived', 'Died'],
    value='All',
    width=filter_width
)

hist_var = pn.widgets.Select(
    name='Histogram Variable',
    options=numeric_cols,
    value='age',
    width=filter_width
)

hist_bins = pn.widgets.IntSlider(
    name='Histogram Bins',
    start=5,
    end=30,
    value=15
)

box_var = pn.widgets.Select(
    name='Boxplot Variable',
    options=numeric_cols,
    value='ejection_fraction',
    width=filter_width
)

scatter_x = pn.widgets.Select(
    name='Scatterplot X Variable',
    options=numeric_cols,
    value='serum_creatinine',
    width=filter_width
)

scatter_y = pn.widgets.Select(
    name='Scatterplot Y Variable',
    options=numeric_cols,
    value='serum_sodium',
    width=filter_width
)

heatmap_vars = pn.widgets.MultiSelect(
    name='Heatmap Variables',
    options=numeric_cols,
    value=numeric_cols,
    size=7
)
```

# Functions

```python

def build_title(base_title:str, sex_value:str, outcome_value:str) -> str:
    """
    Dynamically build the title based on the selected filters
    """
    filters = []
    if sex_value != 'All':
        filters.append(sex_value)
    if outcome_value != 'All':
        filters.append(outcome_value)
    
    if filters:
        return f'{base_title} with filters: {filters}'
    return base_title

def filter_df(df:pd.DataFrame, sex_value:str, outcome_value:str) -> pd.DataFrame:
    """
    Filter the dataframe based on suppled gender value and death outcome value
    """
    filtered_df = df.copy()
    
    if sex_value != 'All':
        filtered_df = filtered_df[filtered_df['sex_label']==sex_value]
    
    if outcome_value != 'All':
        filtered_df = filtered_df[filtered_df['DEATH_EVENT_LABEL']==outcome_value]
    
    return filtered_df

@pn.depends(hist_var, hist_bins, sex_filter, outcome_filter)
def make_histogram(variable:str, bins:int, sex_value:str, outcome_value:str):
    """
    Make a histogram based on the input variable, bins, kde, sex value, and outcome value
    """
    filtered_df = filter_df(df, sex_value, outcome_value)
    title = build_title(f'Histogram of {variable}', sex_value, outcome_value)

    fig = px.histogram(
        filtered_df,
        x = variable,
        nbins = bins,
        title=title
    )

    fig.update_layout(height=400)

    return pn.pane.Plotly(fig, sizing_mode='stretch_width')

@pn.depends(box_var, sex_filter, outcome_filter)
def make_boxplot(variable:str, sex_value:str, outcome_value:str):
    """
    """
    filtered_df = filter_df(df, sex_value, outcome_value)
    title = build_title(f'Boxplot of {variable}', sex_value, outcome_value)

    fig = px.box(
        filtered_df,
        x = 'DEATH_EVENT_LABEL',
        y = variable,
        title = title
    )
    fig.update_layout(height=400)

    return pn.pane.Plotly(fig, sizing_mode='stretch_width')

@pn.depends(scatter_x, scatter_y, sex_filter, outcome_filter)
def make_scatterplot(x_var:str, y_var:str, sex_value:str, outcome_value:str):
    """
    """
    filtered_df = filter_df(df, sex_value, outcome_value)
    title = build_title(f'Scatterplot of {x_var} vs {y_var}', sex_value, outcome_value)

    fig = px.scatter(
        filtered_df,
        x=x_var,
        y=y_var,
        hover_data=[
            'sex_label',
            'smoking_label',
            'diabetes_label',
            'high_blood_pressure_label'
        ],
        title=title
    )

    fig.update_layout(height=400)

    return pn.pane.Plotly(fig, sizing_mode='stretch_width')

@pn.depends(heatmap_vars, sex_filter, outcome_filter)
def make_heatmap(selected_vars, sex_value:str, outcome_value:str):
    """
    """
    filtered_df = filter_df(df, sex_value, outcome_value)

    selected_vars_list = list(selected_vars)
    if len(selected_vars_list) < 2:
        return pn.pane.Markdown("Please select at least 2 variables for the heatmap")
    
    title = build_title('Correlation Heatmap', sex_value, outcome_value)
    corr = filtered_df[selected_vars_list].corr()

    fig = px.imshow(
        corr,
        x=corr.columns,
        y=corr.index,
        aspect='auto',
        text_auto='.2f',
        color_continuous_scale='Viridis',
        title=title
    )

    fig.update_layout(height=400)

    return pn.pane.Plotly(fig, sizing_mode='stretch_width')

@pn.depends(sex_filter, outcome_filter)
def make_table(sex_value:str, outcome_value:str):
    """
    """
    filtered_df = filter_df(df, sex_value, outcome_value)

    return pn.widgets.Tabulator(filtered_df, pagination='local', page_size=10, sizing_mode='stretch_width')
```

# Bind functions to widgets and create Views

```python
histogram_view = pn.Column(
    hist_var,
    hist_bins,
    make_histogram,
    sizing_mode='stretch_width'
)

boxplot_view = pn.Column(
    box_var,
    make_boxplot,
    sizing_mode='stretch_width'
)

scatter_view = pn.Column(
    scatter_x,
    scatter_y,
    make_scatterplot,
    sizing_mode='stretch_width'
)

heatmap_view = pn.Column(
    heatmap_vars,
    make_heatmap,
    sizing_mode='stretch_width'
)

table_view = pn.Column(
    make_table,
    sizing_mode='stretch_width'
)
```

```python
filter_controls = pn.WidgetBox(
    'FILTERS',
    sex_filter,
    outcome_filter,
    width=filter_controls_width,
    sizing_mode='fixed'
)

tabs = pn.Tabs(
    ('Histogram', histogram_view),
    ('Box Plot', boxplot_view),
    ('Scatterplot', scatter_view),
    ('Heatmap', heatmap_view),
    ('Data Table', table_view),
    sizing_mode='stretch_width'
)

dashboard = pn.Row(
    filter_controls,
    tabs,
    sizing_mode='stretch_width'
)
```

# Display Dashboard

```python
# open this in the browser because it doesn't work in VS Code
dashboard.show(port=8080, open=True, threaded=True)
```
